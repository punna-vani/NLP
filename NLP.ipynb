# porter stemmer

```python
from nltk.stem import PorterStemmer
pst=PorterStemmer()
```

```python
pst.stem('using')
```

```python
pst.stem('husbandary')
```

```python
words_to_stem=['write','wrote','eat','eating','think','thinking']
for words in words_to_stem:
    print(words+ ':' +pst.stem(words))
```

```python
from nltk.stem import LancasterStemmer
lancaster = LancasterStemmer()
words_to_stem = ['write','wrote','eat','eating','think','thinking']

for word in words_to_stem:
    stemmed_word = lancaster.stem(word)
    print(word + ': ' + stemmed_word)
```

```python
from nltk.stem import SnowballStemmer
sbst=SnowballStemmer('english')
for words in words_to_stem:
    print(words+':'+sbst.stem(words))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text=("Rice,Wheat,Corn/Maize,foodgrains,foodcrop,foodcrops,vegetables, ,Rize,Potatoes,Tomatoes,Soybeans,Barley,Cassava,Cabbage,Apples,Barley,OatsMillet,Sorghum,Rye,Quinoa")
text
```

```python
wordcloud=WordCloud(width=500,height=600,margin=0).generate(text)
```

```python
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.margins(x=0,y=0)
plt.show()
```

```python
wordcloud=WordCloud(width=300,height=600,margin=0).generate(text)
plt.imshow(wordcloud,interpolation='gaussian')
plt.axis("off")
plt.margins(x=0,y=0)
plt.show()
```

```python
wordcloud=WordCloud(width=1000,height=600,margin=0).generate(text)
plt.imshow(wordcloud,interpolation='quadric')
plt.axis("off")
plt.margins(x=0,y=0)
plt.show()
```

```python
wordcloud=WordCloud(width=1000,height=500,margin=0).generate(text)
plt.imshow(wordcloud,interpolation='bicubic')
plt.axis("off")
plt.margins(x=1,y=2)
plt.show()
```
